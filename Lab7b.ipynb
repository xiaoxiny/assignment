>Explain the dierences that the bottom topography makes.

**A**:The presence of the Corilois force in the shallow water equations causes the anti-cyclonic fluid motion. The magnitude of the surface height depends on the sbolute depth and the slope of bottom. As the water depth decreases due to bottom topography, the surface height increases and the velocity dereases. 

Take the **'Sin function'** topography (the cell below) as an example. Comparing to the flat bottom one, we can see eta increase from 0.13 to 0.58, while the maxium of u decrease from 0.03. Since the change of bottom causes $\nabla$(H$\times$ u) to change more rapidly, the amplitude of the wave (Eta) will change greater than the flat bottom one, which could also be explained by the equation below.

$$\frac{\partial h}{\partial t} + \frac{\partial (H\times u)}{\partial x} + \frac{\partial (H \times v)}{\partial y} = 0$$ 

The **Gaussian function** one simulates the volcanos in the deep sea. It also agrees the fact that if the bottom topography changes significantly, there would be a huge change in $\nabla$(H$\times$ u) and consequently, the the amplitude of the wave will address to reflect the bottom. (in our case, the surface height is sharpened.)
