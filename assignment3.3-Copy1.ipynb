```python
a=np.array([[3,-2],[-4,5]])
print a
print 'cond. number: %s'%np.linalg.cond(a)
```

```python
##find cond. number

def cond(N):
    n=N+1
    matrix=np.empty([n,n])
    matrix[:,:]=0
    for i in range (n):
            matrix[i,i]=-2;
            if i!=0:
                matrix[i,i-1]=1;
            if i!=n-1:
                matrix[i,i+1]=1;
    matrix[0,1]=2
    matrix[n-1,n-2]=2
    if n%2==0:
        matrix[(n)/2,:]=0
        matrix[(n)/2,(n)/2]=1
    else:
        matrix[(n-1)/2,:]=0
        matrix[(n-1)/2,(n-1)/2]=1   
    #print matrix.shape
    #print matrix
    a=np.linalg.cond(matrix)
    #print a
    return a,
```

```python
##find cond. number with boundary cond. substituted
def cond2(N):
    n=N+1
    matrix=np.empty([n,n])
    matrix[:,:]=0
    for i in range (n):
            matrix[i,i]=-2;
            if i!=0:
                matrix[i,i-1]=1;
            if i!=n-1:
                matrix[i,i+1]=1;
    matrix[0,1]=2
    matrix[n-1,:]=1
    #print matrix.shape
    #print matrix
    a=np.linalg.cond(matrix)
    #print n 
    return a,
```

```python
cond2(10)
```

```python
#plot
fig = plt.figure(figsize=(20,10))
for i in range(5,50):

    ax1=plt.subplot(131)
    ax1.scatter(i,cond(i),color='orange',marker='^',alpha=0.8,label='Matrix A1 (u(1/2)=c)')
    plt.xlabel('Matrix size')
    plt.ylabel('Condition number')
    plt.ylim([0,3000])
    ax2=plt.subplot(132)
    ax2.scatter(i,cond2(i),color='red',marker='*',alpha=0.8,label='Matrix A2 (sum=c)')
    plt.xlabel('Matrix size')
    plt.ylabel('Condition number')
    plt.ylim([0,3000])
    ax3=plt.subplot(133)
    ax3.scatter(i,cond(i)[0]-cond2(i)[0],color='green',marker='o',alpha=0.4,label='difference')
    plt.xlabel('Matrix size')
    plt.ylabel('Difference between two condition numbers')
    ax1.hold(True)
    ax2.hold(True)
    ax3.hold(True)
    if i==5:
        ax1.legend(loc='lower right', shadow=True)
        ax2.legend(loc='lower right', shadow=True)
        ax3.legend(loc='upper right', shadow=True)
ax1.grid(True)
ax2.grid(True)
ax3.grid(True)
plt.show()
```
