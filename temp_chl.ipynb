```python
import csv
filename = 'ARCSS-PP.csv'
lats, lons,year = [], [],[]

# Read through the entire file, skip the first line,
#  and pull out just the lats and lons.
with open(filename) as f:
    # Create a csv reader object.
    reader = csv.reader(f)
    
    # Ignore the header row.
    next(reader)
    
    # Store the latitudes and longitudes in the appropriate lists.
    for row in reader:
        lats.append(float(row[5]))
        lons.append(float(row[6]))
        year.append(float(row[1]))
```

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10,10))
my_map = Basemap(lon_0=0, boundinglat=60, projection='npstere',resolution = 'h',round=True)

x,y=my_map(lons,lats) 
my_map.bluemarble()
my_map.drawcoastlines()
my_map.drawcountries()
#my_map.fillcontinents(color='coral')
my_map.drawmapboundary()

my_map.scatter(x,y,c=year)
my_map.drawmeridians(np.arange(0, 360, 30))
my_map.drawparallels(np.arange(-90, 90, 30))
plt.show()
```

```python
plt.figure(figsize=(10,10))
my_map = Basemap(lon_0=0, boundinglat=60, projection='npstere',resolution = 'h',round=True)

x,y=my_map(lons,lats) 
my_map.bluemarble()
my_map.drawcoastlines()
my_map.drawcountries()
#my_map.fillcontinents(color='coral')
my_map.drawmapboundary()
for i in range (len(x)):
    if year[i]==1987 or year[i]==1986 or year[i]==1985:
        my_map.scatter(x[i],y[i],c=year[i]*10)
my_map.drawmeridians(np.arange(0, 360, 30))
my_map.drawparallels(np.arange(-90, 90, 30))
plt.show()
```

```python
plt.figure(figsize=(10,10))
my_map = Basemap(lon_0=0, boundinglat=60, projection='npstere',resolution = 'h',round=True)

x,y=my_map(lons,lats) 
my_map.bluemarble()
my_map.drawcoastlines()
my_map.drawcountries()
my_map.drawmapboundary()

for i in range (len(x)):
    if year[i]<1987:
        my_map.scatter(x[i],y[i],c=year[i]/1000)
my_map.drawmeridians(np.arange(0, 360, 30))
my_map.drawparallels(np.arange(-90, 90, 30))
plt.show()
```
